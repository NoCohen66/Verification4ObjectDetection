```python
import pandas as pd
import plotly.express as px
```

# LARD

```python
df = pd.read_csv("CGT_comp_LARD/1_iou_calculations.csv", index_col=0,  converters={"lb_box":pd.eval,  
                                                                                   'ub_box':pd.eval,
                                                                                   'IoU_extension':pd.eval})
df['IoU_extension_min'] = df['IoU_extension'].apply(lambda x: x[0])
df = df[['eps', 'perturbation', 'lb_box', 'ub_box', 'IoU_extension_min']]
df = df[df["perturbation"] !=  'contrast']
```

```python
import numpy as np
np.unique(df.perturbation)
```

```python
df
```

```python
fig = px.bar(df, x='eps', y='IoU_extension_min', color='perturbation', barmode="group")
fig.show()
```



<div class="alert alert-block alert-info"> <b>CONCLUSION</b> For this dataset, the comparison of produced bounds with bounded inputs shows similarity between 'CGT brightness' and 'whitenoise.' Furthermore, 'CGT brightness' and 'CGT brightness clamp' are similar due to the absence of a black monochromatic background. </div>

# MNIST

```python
df = pd.read_csv("CGT_comp/0_iou_calculations.csv", index_col=0,  converters={"lb_box":pd.eval,  'ub_box':pd.eval,'IoU_extension':pd.eval})
df['IoU_extension_min'] = df['IoU_extension'].apply(lambda x: x[0])
df = df[['eps', 'perturbation', 'lb_box', 'ub_box', 'IoU_extension_min']]
df[df["eps"] == 0.0014]
```

<div class="alert alert-block alert-info"> <b>CONCLUSION</b> Similar results are observed between 'CGT brightness' and 'brightness' as concluded from LARD dataset. The CGT authors' implementation of brightness is not correct.</div>



<div class="alert alert-block alert-success"> <b>REMARK</b> However, for this dataset, there is a biasing factor in the results. Since MNIST features a black monochromatic background, failing to apply a clamp allows for the presence of negative pixels, adversely affecting the model. Therefore, it would be beneficial for future implementations to include a clamping mechanism as they currently do.</div>
