```python
import plotly.express as px

def create_box_plot(df, columns):
    # Melt the dataframe to go from wide to long format
    df_long = df.melt(value_vars=columns, var_name='Time', value_name='Value')
    
    # Create the box plot
    fig = px.box(df_long, x='Time', y='Value', color='Time', 
                 notched=True,  # shows the confidence interval around the median
                 title='Distribution of Times')
    
    # Customize the layout for better appearance
    fig.update_layout(
        yaxis_title='Time (s)',  # Replace with your actual units
        xaxis_title='IoU calculation methods (elapsed time)',
        template='plotly_white',  # A clean template
        showlegend=False  # Hide the legend if not necessary
    )
    
    # Show the plot
    fig.show()

# Example usage with a DataFrame 'df' and the columns you want to plot
create_box_plot(df, ["tmps_vanilla",	"tmps_extension",	"tmps_greedy"])
```
