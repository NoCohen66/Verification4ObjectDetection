## <a id="without"></a> II -  Pipeline without perturbation

### I-B. Object detection basics

**Bounding box** is a rectangle encapsulating the object of interest.

**Ground truth** is the bounding box accurately placed around an object within an image, known to be true.

**Intersection over Union metric** can evaluate how similar two bounding boxes are.
    


```python
# Importation
import os 
import sys
import torch
import numpy as np
from contextlib import closing
import pickle as pkl
from PIL import Image, ImageDraw
sys.path.append(os.path.join(os.getcwd(),'..'))
from solver.model.LARD.LARD import Neural_network_LARD, Neural_network_LARD_BrightnessContrast
LARD_model_weights = '../solver/model/LARD/weights.pt'
LARD_dataset = '../solver/data/LARD/test.pkl'
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")



# Load LARD model and data
model_torch_load  = torch.load(LARD_model_weights, torch.device('cpu')) # change if GPU
torch_model = Neural_network_LARD()
dict_val = {}
for name, param in zip(torch_model.state_dict().keys(),
                        model_torch_load.state_dict().values()):
            dict_val[name] = param

torch_model.load_state_dict(dict_val)
with closing((open(LARD_dataset, 'rb'))) as f:
                    dico_dataset = pkl.load(f)        
X_test = torch.Tensor(dico_dataset['x_train']).to(device)
Y_test = torch.Tensor(dico_dataset['y_train']).to(device)


# An inference on an image X_0
index=-4
X_0 = X_test[index][None]/255
gt_box = Y_test[index]*256

# Drawing bounding boxes 
ymin, xmin, ymax, xmax  = gt_box.detach().numpy()
pred1_ymin, pred1_xmin, pred1_ymax, pred1_xmax = torch_model(X_0)[0].detach().numpy()
pil_image = Image.fromarray(X_test[index].byte().permute(1,2,0).numpy(), mode="RGB")
draw = ImageDraw.Draw(pil_image)
draw.rectangle([ymin, xmin, ymax, xmax], outline="green", width=3)
draw.rectangle([pred1_ymin, pred1_xmin, pred1_ymax, pred1_xmax], outline="red")
pil_image
```

```python
# Importation
from iou_calculator.bounding_box_utils.bounding_box import Hyperrectangle
from iou_calculator.bounding_box_utils.bounding_box_interval_values import Hyperrectangle_interval
from iou_calculator.bounding_box_utils.interval import Interval
from iou_calculator.IoU import IoU

# IoU calculation
ground_truth_box = Hyperrectangle(x_bl=xmin,x_tr=xmax, y_bl=ymin, y_tr=ymax)
predicted_box = Hyperrectangle_interval(x_1=Interval(pred1_xmin), x_2=Interval(pred1_xmax), y_1=Interval(pred1_ymin), y_2=Interval(pred1_ymax))
iou_value = IoU(predicted_box,ground_truth_box).iou_simple()


print(f"The IoU value between the predicted box and the ground truth is: {iou_value}")
```

<div class="alert-info">
    The prediction is indicated in red, while the ground truth is shown in green.  
    The intersection over union (IoU) value approaches 1, signifying a high degree of accuracy in the prediction.
</div>

## <a id="with"></a> II -  Pipeline with perturbation

<img src="../results/images/with_perturbations.png" width="800" />

The processing pipeline for the analysis of an input image is structured as follows: 
- Initially,  we compute the abstract domain for an input image: it represents an input image under a range of perturbation.
- Then, we compute the set of reachable bounding boxes, utilizing the [auto-LiRPA solver](https://github.com/Verified-Intelligence/auto_LiRPA).
- Finally, the IoU metric is calculated with our two methods: Vanilla_IoU and Optimal_IoU.


### II-A. Whitenoise perturbation

```python
# Import code to apply a whitenoise perturbation
from solver.perturbation import bound_whitenoise

# Select intensity (you can modify this value to see the impact)
perturbation_intensity = 0.0009

# Computes the set of reachables bounding boxes
lb_box_wn, ub_box_wn = bound_whitenoise(torch_model, X_0, perturbation_intensity, method="CROWN")
predicted_box_wn = Hyperrectangle_interval(x_1=Interval(lb_box_wn[1],ub_box_wn[1]), x_2=Interval(lb_box_wn[3],ub_box_wn[3]), y_1=Interval(lb_box_wn[0],ub_box_wn[0]), y_2=Interval(lb_box_wn[2],ub_box_wn[2]))
```

The outer bound of the set of reachables bouding boxes in in red and the inner bound of the set of reachables bouding boxes is in orange.

```python
pred1_ymin, pred1_xmin, pred1_ymax, pred1_xmax = lb_box_wn
pred2_ymin, pred2_xmin, pred2_ymax, pred2_xmax = ub_box_wn

pil_image = Image.fromarray(X_test[index].byte().permute(1,2,0).numpy(), mode="RGB")
draw = ImageDraw.Draw(pil_image)
draw.rectangle([ymin, xmin, ymax, xmax], outline="green", width=3)
draw.rectangle([pred1_ymin, pred1_xmin, pred1_ymax, pred1_xmax], outline="red")
draw.rectangle([pred2_ymin, pred2_xmin, pred2_ymax, pred2_xmax], outline="orange")
pil_image
```

### <a id="IoU"></a> II.B -  IoU_Vanilla

The fundamental idea is to enhance the traditional arithmetic system by employing closed intervals as basic data types, rather than single point values, on each primitive of the IoU function. Indeed, IoU, can be formulated as a combination of so-called primitive functions: minimum, maximum, addition, subtraction, multiplication and division by a positive scalar.

```python
iou_value = IoU(predicted_box_wn,ground_truth_box).iou()
 

print("The range of IoU_vanilla is:", iou_value['IoU_vanilla'], "computed in", iou_value['tmps_vanilla'], "seconds.")
```

### II.C - IoU_optimal 

Our objective is to identify, from the set of reachable bounding boxes, the bounding box that maximizes the IoU value as well as the one that minimizes it. To accomplish this, we employ partial derivatives to locate points of local maximum and minimum. This method allows us to precisely determine the bounding boxes that yield the highest and lowest IoU values.

```python
print("The range of IoU_optimal is:", iou_value['IoU_extension'], "computed in", iou_value['tmps_extension'], "seconds.")
```

As expected, the bounds determined by IoU_Optimal method are more precise than the one with Vanilla_IoU.

### III-A. Brightness and Contrast perturbation


To apply brightness and contrast perturbations effectively, we utilize the perturbation coefficient, alpha. 
        For brightness adjustments, alpha is added, and for contrast, it is multiplied.
        The tool auto-lirpa is particularly compatible with neural networks designed to take alpha as the sole input during their forward operation. 
        To accommodate this, we have developed a neural network that depends on the image and includes a linear layer. 
        This layer is responsible for performing either the addition or multiplication of the image with alpha as the input.


```python
from solver.model.LARD.LARD import  Neural_network_LARD_BrightnessContrast
from solver.perturbation import bound_brightness_LARD

model_bri = Neural_network_LARD_BrightnessContrast()
bri_config = dict_val
bri_config['linear_perturbation.weight']=torch.Tensor(np.zeros((3*256*256, 1), dtype='float32'))
bri_config['linear_perturbation.bias']=torch.Tensor(np.zeros((3*256*256,), dtype='float32'))
model_bri.load_state_dict(bri_config)
```

```python
# Computes the set of reachables bounding boxes

lb_box_bri, ub_box_bri = bound_brightness_LARD(model_bri, X_0, perturbation_intensity, method="CROWN")

predicted_box_bri = Hyperrectangle_interval(x_1=Interval(lb_box_bri[1],ub_box_bri[1]), x_2=Interval(lb_box_bri[3],ub_box_bri[3]), y_1=Interval(lb_box_bri[0],ub_box_bri[0]), y_2=Interval(lb_box_bri[2],ub_box_bri[2]))
```

### III-B. IoU_Vanilla

```python
iou_value_bri = IoU(predicted_box_bri,ground_truth_box).iou()
 

print("The range of IoU_vanilla is:", iou_value_bri['IoU_vanilla'], "computed in", iou_value_bri['tmps_vanilla'], "seconds.")
```

### III-B. IoU_Optimal

```python
print("The range of IoU_optimal is:", iou_value_bri['IoU_extension'], "computed in", iou_value_bri['tmps_extension'], "seconds.")
```

As expected, when perturbations involve uniform perturbation across pixels (for instance: brightness), the resulting IoU is more accurate and tighter compared to scenarios where the perturbations affect pixel non-uniformly (for instance: whitenoise).
